## Data Cleaning

```python
df_train.info()
```

```python
df_train1 = df_train.dropna(thresh = 50)
df_train1 = df_train1.dropna(axis =1, thresh = 200)
df_train1 = df_train1[df_train1["sii"].notnull()]
print(df_train1.info())
print(df_train1.shape)


```

```python
def fillMissingValues(df):
    df = df.reset_index(drop= True)
    float_cols = df.select_dtypes(include= ["float"]).columns
    float_cols = [col for col in float_cols if df[col].isnull().any()]
    sii_value = [int(x) for x in df["sii"].unique()]
#     for sii in [0, 1, 2, 3]:
#         print(f"Data for sii = {sii}:\n", df[df["sii"] == sii][float_cols].describe())
    print(df[df["sii"] == 0.0]["CGAS-CGAS_Score"].mean())
    for i in float_cols:
        for sii in sii_value:
            col_mean = float(df[df["sii"] == sii][i].mean())
            col_median = float(df[df["sii"] == sii][i].median())
            print(f"{i} sii : {sii} mean : {col_mean}  median : {col_median}")
            df[i] = df[i][df["sii"] == sii].fillna(col_median)
        
fillMissingValues(df_train1.copy())
```
